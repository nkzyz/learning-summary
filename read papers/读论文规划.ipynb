1 目的    
分析领域现有技术，尝试构思一些创新方案，做出一些初步结果

2做法    
1）看领域核心论文和老文献。总结对比，整理成综述。       
核心文献：关注的实验室发表的文献；顶刊上的，引用量不太多的文献；被多次提及的文献，一般是早期文献。     
老文献：从机器学习的发展历程，选择了解可能和我相关的、被证明有效的技术 以及后来对它的解释。并关注对应的作者。(看看有没有介绍发展过程的书和课程)       
关注的技术方向：预训练大模型微调，多模态图像表示      
1.1）由 新意度 有效性 判断读的程度     
关注 应用场景的介绍 当前技术思路(在方法部分，关注对作用的介绍) 思路来源（刻意学习模仿这部分 ）和效果，思考可能的局限性(观察结果 思考存在的问题)，哪些部分有用应该保留(消融实验)，哪些能变换一下(更准确，更便宜，更安全，更大规模)。分析文章档次。      
很多作者自己也不清楚，因此不要求 理解技术路线在逻辑上的可行性。判断什么技术部分要理解清楚，挑选关键的部分理解，帮助理解进展。       
1.2）复现、测试和改进 两种类型：很有效且自己用得上的；方法思路很新，自己没尝试过，可能以后能拓展出更好结果的。发现局限性，解决问题。（关注技术的排列组合 和改进）       
1.3）看该作者 或引用的关键技术作者 对应期刊是否要关注。稍微看一下作为对比的其他模型，以及对应的课题组。       
2）过程中，如果有一个新的想法，一边查文献，一边就开始做。而不是单独查文献。      


3 任务进度控制和截止时间         
  1）先预计做应该要多久，总时间分为：构思的时间(包括生活中的零碎时间)，做的时间，多余的时间（2天左右）              
  2) 可接受的最低版本是：测试技术有效性，分析思路局限性，想出可能的设计。   
  3）如果遇到了意外的问题，没有按时复现出来，只能吸取教训 以后分配更多时间给意外事件，并且每天按需要增加时间。
